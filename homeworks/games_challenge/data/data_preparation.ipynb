```python
import pandas as pd
```

```python
output_steamspy = pd.read_csv('output_steamspy.csv')
steamstore = pd.read_csv('steam_store_data_2024.csv')
test = pd.read_csv('test.csv')
train = pd.read_csv('train.csv')

output_steamspy
```

```python
# different appids in test and train
test_appids = set(test['app_id'])
train_appids = set(train['app_id'])

print(f"Number of unique appids in test: {len(test_appids)}")
print(f"Number of unique appids in train: {len(train_appids)}")
```

```python
#check if there are any overlaps in output_steamspy.name and steamstore.title
overlap_names = set(output_steamspy['name']).intersection(set(steamstore['title']))
print(f"Number of overlapping names between output_steamspy and steamstore: {len(overlap_names)}")

# check if there are any overlaps test_appids, train_appids, output_steamspy.appid
print(f"Total different appids in train and test: {len(test_appids.union(train_appids))}")
overlap_appids = test_appids.union(train_appids).intersection(set(output_steamspy['appid']))
print(f"Number of overlapping appids between test, train, and output_steamspy: {len(overlap_appids)}")

#ids that do not overlap
non_overlap_appids = test_appids.union(train_appids).difference(set(output_steamspy['appid']))
print(non_overlap_appids)
```
